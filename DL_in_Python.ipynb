```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
import matplotlib.pyplot as plt
import random
from sklearn.model_selection import train_test_split

(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
assert x_train.shape == (50000, 32, 32, 3)
assert x_test.shape == (10000, 32, 32, 3)
assert y_train.shape == (50000, 1)
assert y_test.shape == (10000, 1)

# Code to pick random numbers for the index of the image
r1 = random.randint(1,50000)
r2 = random.randint(1,50000)
r3 = random.randint(1,50000)
r4 = random.randint(1,50000)
r5 = random.randint(1,50000)
r6 = random.randint(1,50000)
r7 = random.randint(1,50000)
r8 = random.randint(1,50000)
r9 = random.randint(1,50000)

# Identifying the category that the selected photo is in
def output_dim(input):
  if input <= 5000:
    print('airplane')
  elif input <= 10000:
    print('automoble')
  elif input <= 15000:
    print('bird')
  elif input <= 20000:
    print('cat')
  elif input <= 25000:
    print('deer')
  elif input <= 30000:
    print('dog')
  elif input <= 35000:
    print('frog')
  elif input <= 40000:
    print('horse')
  elif input <= 45000:
    print('ship')
  elif input <= 50000:
    print('truck')

# Calls to the output_dim function to get the category that the photo is in  
o1 = output_dim(r1)
output_dim(r2)
output_dim(r3)
output_dim(r4)
output_dim(r5)
output_dim(r6)
output_dim(r7)
output_dim(r8)
output_dim(r9)

# Code to create a 3 x 3 plot
f, axarr = plt.subplots(3,3)
axarr[0,0].imshow(x_train[r1])
axarr[0,1].imshow(x_train[r2])
axarr[0,2].imshow(x_train[r3])
axarr[1,0].imshow(x_train[r4])
axarr[1,1].imshow(x_train[r5])
axarr[1,2].imshow(x_train[r6])
axarr[2,0].imshow(x_train[r7])
axarr[2,1].imshow(x_train[r8])
axarr[2,2].imshow(x_train[r9])

print(' ')

#Printing the index number from the dataset
print('Input photo 1: ', r1)
print('Input photo 2: ', r2)
print('Input photo 3: ', r3)
print('Input photo 4: ', r4)
print('Input photo 5: ', r5)
print('Input photo 6: ', r6)
print('Input photo 7: ', r7)
print('Input photo 8: ', r8)
print('Input photo 9: ', r9)


print('Input train dim:', x_train.shape) # Input values
print('Output train dim: ', y_train.shape)  # Output values

#Normalizing the data using min-max normalization
X_std = (((r1) - 1) / (50000 - 1))
X_scaled = X_std * (50000 - 1) + 1

print('')
print(X_std)
print(X_scaled)

# Using the train_test_split_function package from sklearn
print('The following is the result of the test_train_split function in sklearn')
X_train, X_test, Y_train, Y_test = train_test_split((x_train, x_test), (y_train, y_test), test_size=0.20, random_state=42)
display(X_train)
display(X_test)
display(Y_train)
display(Y_test)

```
